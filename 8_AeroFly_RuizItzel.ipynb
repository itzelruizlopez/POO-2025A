```python
from datetime import datetime

# Clase para representar a un pasajero
class Pasajero:
    def __init__(self, nombre, numero_pasaporte, contacto):
        self.nombre = nombre
        self.numero_pasaporte = numero_pasaporte
        self.contacto = contacto

    def mostrar_info(self):
        return f"{self.nombre} ({self.numero_pasaporte}) - {self.contacto}"

# Clase para representar un vuelo
class Vuelo:
    def __init__(self, numero_vuelo, destino, fecha_salida, hora_salida, capacidad):
        self.numero_vuelo = numero_vuelo
        self.destino = destino
        self.fecha_salida = fecha_salida
        self.hora_salida = hora_salida
        self.capacidad = capacidad
        self.asientos_disponibles = capacidad
        self.pasajeros = []

    def reservar_asiento(self, pasajero):
        if self.asientos_disponibles > 0:
            self.pasajeros.append(pasajero)
            self.asientos_disponibles -= 1
            return True
        return False

    def cancelar_reserva(self, pasajero):
        if pasajero in self.pasajeros:
            self.pasajeros.remove(pasajero)
            self.asientos_disponibles += 1
            return True
        return False

    def mostrar_info(self):
        return f"Vuelo {self.numero_vuelo} a {self.destino} - {self.fecha_salida} {self.hora_salida}"

# Clase para representar una reserva
class Reserva:
    def __init__(self, vuelo, pasajero, numero_asientos):
        self.vuelo = vuelo
        self.pasajero = pasajero
        self.numero_asientos = numero_asientos
        self.codigo_reserva = f"RES-{self.vuelo.numero_vuelo}-{self.pasajero.numero_pasaporte}"

    def mostrar_info(self):
        return f"Reserva {self.codigo_reserva}: {self.pasajero.mostrar_info()} - {self.vuelo.mostrar_info()}"

# Clase para representar la aerolínea
class Aerolinea:
    def __init__(self, nombre):
        self.nombre = nombre
        self.destinos = []
        self.vuelos = []

    def agregar_destino(self, destino):
        if destino not in self.destinos:
            self.destinos.append(destino)

    def eliminar_destino(self, destino):
        if destino in self.destinos:
            self.destinos.remove(destino)

    def mostrar_destinos(self):
        return self.destinos

    def crear_vuelo(self, numero_vuelo, destino, fecha_salida, hora_salida, capacidad):
        if destino in self.destinos:
            vuelo = Vuelo(numero_vuelo, destino, fecha_salida, hora_salida, capacidad)
            self.vuelos.append(vuelo)
            return vuelo
        return None

    def buscar_vuelo(self, destino, fecha_salida, hora_salida):
        for vuelo in self.vuelos:
            if vuelo.destino == destino and vuelo.fecha_salida == fecha_salida and vuelo.hora_salida == hora_salida:
                return vuelo
        return None

# Función para interactuar con el sistema
def sistema_reservas():
    eurofly = Aerolinea("EuroFly")
    eurofly.agregar_destino("Madrid")
    eurofly.agregar_destino("Barcelona")
    eurofly.agregar_destino("Roma")

    while True:
        print("\n--- Sistema de Reservas EuroFly ---")
        print("1. Mostrar destinos disponibles")
        print("2. Crear vuelo")
        print("3. Buscar vuelo")
        print("4. Reservar asiento")
        print("5. Cancelar reserva")
        print("6. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            destinos = eurofly.mostrar_destinos()
            print("\nDestinos disponibles:")
            for destino in destinos:
                print(f"- {destino}")

        elif opcion == "2":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            if destino not in eurofly.destinos:
                print("Destino no válido.")
                continue
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            capacidad = int(input("Capacidad del vuelo: "))
            vuelo = eurofly.crear_vuelo(numero_vuelo, destino, fecha_salida, hora_salida, capacidad)
            if vuelo:
                print(f"Vuelo {numero_vuelo} creado con éxito.")
            else:
                print("No se pudo crear el vuelo.")

        elif opcion == "3":
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                print(f"Vuelo encontrado: {vuelo.mostrar_info()}")
            else:
                print("No se encontró el vuelo.")

        elif opcion == "4":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                nombre = input("Nombre del pasajero: ")
                numero_pasaporte = input("Número de pasaporte: ")
                contacto = input("Información de contacto: ")
                pasajero = Pasajero(nombre, numero_pasaporte, contacto)
                if vuelo.reservar_asiento(pasajero):
                    reserva = Reserva(vuelo, pasajero, 1)
                    print(f"Reserva exitosa: {reserva.mostrar_info()}")
                else:
                    print("No hay asientos disponibles.")
            else:
                print("No se encontró el vuelo.")

        elif opcion == "5":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                nombre = input("Nombre del pasajero: ")
                numero_pasaporte = input("Número de pasaporte: ")
                pasajero = Pasajero(nombre, numero_pasaporte, "")
                if vuelo.cancelar_reserva(pasajero):
                    print("Reserva cancelada con éxito.")
from datetime import datetime

# Clase para representar a un pasajero
class Pasajero:
    def __init__(self, nombre, numero_pasaporte, contacto):
        self.nombre = nombre
        self.numero_pasaporte = numero_pasaporte
        self.contacto = contacto

    def mostrar_info(self):
        return f"{self.nombre} ({self.numero_pasaporte}) - {self.contacto}"

# Clase para representar un vuelo
class Vuelo:
    def __init__(self, numero_vuelo, destino, fecha_salida, hora_salida, capacidad):
        self.numero_vuelo = numero_vuelo
        self.destino = destino
        self.fecha_salida = fecha_salida
        self.hora_salida = hora_salida
        self.capacidad = capacidad
        self.asientos_disponibles = capacidad
        self.pasajeros = []

    def reservar_asiento(self, pasajero):
        if self.asientos_disponibles > 0:
            self.pasajeros.append(pasajero)
            self.asientos_disponibles -= 1
            return True
        return False

    def cancelar_reserva(self, pasajero):
        if pasajero in self.pasajeros:
            self.pasajeros.remove(pasajero)
            self.asientos_disponibles += 1
            return True
        return False

    def mostrar_info(self):
        return f"Vuelo {self.numero_vuelo} a {self.destino} - {self.fecha_salida} {self.hora_salida}"

# Clase para representar una reserva
class Reserva:
    def __init__(self, vuelo, pasajero, numero_asientos):
        self.vuelo = vuelo
        self.pasajero = pasajero
        self.numero_asientos = numero_asientos
        self.codigo_reserva = f"RES-{self.vuelo.numero_vuelo}-{self.pasajero.numero_pasaporte}"

    def mostrar_info(self):
        return f"Reserva {self.codigo_reserva}: {self.pasajero.mostrar_info()} - {self.vuelo.mostrar_info()}"

# Clase para representar la aerolínea
class Aerolinea:
    def __init__(self, nombre):
        self.nombre = nombre
        self.destinos = []
        self.vuelos = []

    def agregar_destino(self, destino):
        if destino not in self.destinos:
            self.destinos.append(destino)

    def eliminar_destino(self, destino):
        if destino in self.destinos:
            self.destinos.remove(destino)

    def mostrar_destinos(self):
        return self.destinos

    def crear_vuelo(self, numero_vuelo, destino, fecha_salida, hora_salida, capacidad):
        if destino in self.destinos:
            vuelo = Vuelo(numero_vuelo, destino, fecha_salida, hora_salida, capacidad)
            self.vuelos.append(vuelo)
            return vuelo
        return None

    def buscar_vuelo(self, destino, fecha_salida, hora_salida):
        for vuelo in self.vuelos:
            if vuelo.destino == destino and vuelo.fecha_salida == fecha_salida and vuelo.hora_salida == hora_salida:
                return vuelo
        return None

# Función para interactuar con el sistema
def sistema_reservas():
    eurofly = Aerolinea("EuroFly")
    eurofly.agregar_destino("Madrid")
    eurofly.agregar_destino("Barcelona")
    eurofly.agregar_destino("Roma")

    while True:
        print("\n--- Sistema de Reservas EuroFly ---")
        print("1. Mostrar destinos disponibles")
        print("2. Crear vuelo")
        print("3. Buscar vuelo")
        print("4. Reservar asiento")
        print("5. Cancelar reserva")
        print("6. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            destinos = eurofly.mostrar_destinos()
            print("\nDestinos disponibles:")
            for destino in destinos:
                print(f"- {destino}")

        elif opcion == "2":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            if destino not in eurofly.destinos:
                print("Destino no válido.")
                continue
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            capacidad = int(input("Capacidad del vuelo: "))
            vuelo = eurofly.crear_vuelo(numero_vuelo, destino, fecha_salida, hora_salida, capacidad)
            if vuelo:
                print(f"Vuelo {numero_vuelo} creado con éxito.")
            else:
                print("No se pudo crear el vuelo.")

        elif opcion == "3":
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                print(f"Vuelo encontrado: {vuelo.mostrar_info()}")
            else:
                print("No se encontró el vuelo.")

        elif opcion == "4":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                nombre = input("Nombre del pasajero: ")
                numero_pasaporte = input("Número de pasaporte: ")
                contacto = input("Información de contacto: ")
                pasajero = Pasajero(nombre, numero_pasaporte, contacto)
                if vuelo.reservar_asiento(pasajero):
                    reserva = Reserva(vuelo, pasajero, 1)
                    print(f"Reserva exitosa: {reserva.mostrar_info()}")
                else:
                    print("No hay asientos disponibles.")
            else:
                print("No se encontró el vuelo.")

        elif opcion == "5":
            numero_vuelo = input("Número de vuelo: ")
            destino = input("Destino: ")
            fecha_salida = input("Fecha de salida (YYYY-MM-DD): ")
            hora_salida = input("Hora de salida (HH:MM): ")
            vuelo = eurofly.buscar_vuelo(destino, fecha_salida, hora_salida)
            if vuelo:
                nombre = input("Nombre del pasajero: ")
                numero_pasaporte = input("Número de pasaporte: ")
                pasajero = Pasajero(nombre, numero_pasaporte, "")
                if vuelo.cancelar_reserva(pasajero):
                    print("Reserva cancelada con éxito.")
                else:
                    print("Vuelva a intentar.")
            else:
                print("No se encontró el vuelo.")
        elif opcion == "6":
            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")
```
